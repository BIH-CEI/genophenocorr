# Running a Fisher-Exact or Chi<sup>2</sup> Test

Running the Correlation test is done with the command "run_stats". In this example, we will be testing if there is a significant difference between missense variants and nonsense variants. <br>

```
run_stats(cohort, Fun1, Fun2, extraVar_1, extra_Var2, chi2 = False)
```
cohort = a Cohort class variable with patients to be analyzed. <br><br>
In this test, we want to know whether a patient who has a specific HPO also has either a variant of type "missense" or of type "nonsense" and compare the two against each other. This is done with the function "is_var_type" (See below for further description). In the first test, we want to test for "missense" mutations, and the second test for "nonsense" mutations. This is why the command is set up with the same tests and different variables. <br><br>
chi2 is defaulted to False, meaning a Fisher-Exact test will run unless you set chi2 = True, which it will then run a Chi<sup>2</sup> test. <br><br>
The return of run_stats is a pandas DataFrame of the results. 



````python
result_FE = run_stats(allPatients, is_var_type, is_var_type, 'missense', 'nonsense')
result_FE.head(5)
````

````python
result_chi2 = run_stats(allPatients, is_var_type, is_var_type, 'missense', 'nonsense', chi2 = True)
result_chi2.head(5)
````

# Variant Tests

### Variant Type

**is_var_type** - Along with the examples above, there are other options for types of variants that can be tested, as listed below.<br>
**is_not_var_type** - This is the opposite of is_var_type. Has all of the same types of variants but will only count the patients who do not have that variant type but do and do not have that HPO. 
- missense
- nonsense
- duplication
- deletion
- insertion
- transversion
- transition
- indel


````python
result_not = run_stats(allPatients, is_var_type, is_not_var_type, 'missense', 'missense')
result_not.head(5)
````

### Variant Match
**is_var_match** - Rather than a variant type, this function tests whether a patient has a specific variant. The variable values can either be a Variant Class or a string with the format 'chromosome:start:reference:alternative".<br>
**is_not_var_match** - This is similarly the opposite of is_var_match. It tests whether a patient does not have a specific variant, and takes the same formats as input. 
<br> 
**Example input variables (all are for the same variant):**
- '16:1762843:C:T'
- varcode.Variant(16, 1762843, C, T, pyensembl)
- allPatients.all_patients.get('8').variant.variant


````python
result_match = run_stats(allPatients, is_var_match, is_not_var_match, '16:1762843:C:T', '16:1762843:C:T')
result_match.head(5)
````
